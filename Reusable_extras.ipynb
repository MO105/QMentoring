## Randomizer


```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = []
for i in range(0,99999):
    x = [10,20,30,40,50,60,70,80,90]
    i = random.choice(x)
    y.append(i)
```

```python
# counting frequencies
z = np.array(np.unique(y, return_counts=True)).T
type(z)
```

### checking if the random distributing is really random or not



```python
z[0][0], z[0][1], z[1][0], z[1][1]
a = []
b = []
for i in range(0, len(z)):
    a.append(z[i][0])
    b.append(z[i][1])
```
